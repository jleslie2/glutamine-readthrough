```python
import pyranges as pr
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df_RPKM= pd.read_csv('45864_mRNA_S13_L003_R1_001_qexpr_rpkm.txt', sep='\t', header=None, 
                     names=["gene_id", "length", "footprint_reads", "RPM", "length_kb", "RPKM"])
```

```python
def get_rpkm(df, gene_id):
    return df.loc[df['gene_id'] == gene_id, 'RPKM'].iloc[0]

ade1_RPKM = get_rpkm(df_RPKM, "YAR015W")
ade1_RPKM
```

```python
path = ("/mnt/unallab/jleslie/20250827_ribosome_profiling_snf5_cyc8_ade2/28_29merAnalysis/cyc8analysis/45864_monosome_S9_L003_R1_001_Aligned_unique_sorted_fwd_YAR015W.bed")

gr = pr.read_bed(path)
df = gr.df
```

```python
region_start  = 169374
region_end    = 170292
num_positions = region_end - region_start

df['Score'] = df['Name'].div(float(ade1_RPKM))
df['Score'] = df['Score'].fillna(0)

score_dict_ade1 = {i: 0 for i in range(num_positions)}

for index, row in df.iterrows():
    pos = row['Start'] - region_start
    if 0 <= pos < num_positions:
        score_dict_ade1[pos] = row['Score']

score_list_ade1 = list(score_dict_ade1.values())

df_ade1 = pd.DataFrame({
    "Position": list(score_dict_ade1.keys()),
    "Score":    score_list_ade1
})

#df_ade1.to_csv('041526_45865_normalized_asiteoccupancy_monosomes_ade1.csv', index=False)
```
